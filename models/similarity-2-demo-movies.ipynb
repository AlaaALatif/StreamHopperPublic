### MOVIES

```python
# guidebox data
gb_df = pd.read_csv(data_path/'movies.csv')
meta_df = pd.read_csv(data_path/'movies-meta.csv')
plots_df = pd.read_csv(data_path/'movies-text.csv')
imgs_df = pd.read_csv(data_path/'movies-imgs.csv')
meta_df = load_plot_metadata(meta_df, plots_df)
merged_df = gb_df.merge(meta_df, how='inner', on='imdb_id')
```

```python
starter_meta_df = get_starter_metadata(starter_df, 'movie')
```

```python
x_df = compute_similarity_scores(starter_meta_df, merged_df, imgs_df, 'movie')
```

```python
x_df['avg_score'] = x_df[['rating_score', 'votes_score', 'directors_score',
                          'cast_score', 'writers_score', 'producers_score',
                          'casting directors_score', 'composers_score',
                          'plot_score']].mean(axis=1)
```
